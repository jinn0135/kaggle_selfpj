```python
def cal_dist(a,b,c, x1,y1):
    # ax+by+c=0, (x1,y1) -> dist= |ax1+by1+c|/(a**2+b**2)**(1/2)
    return abs(a*x1+b*y1+c)/np.sqrt(a**2+b**2)

train_check = train.iloc[:,1:].copy()
train_check['medinc/p'] = train_check['MedInc']/train_check['Population']
train_check['br/r'] = train_check['AveBedrms']/train_check['AveRooms']
train_check['dist'] = cal_dist(a,-1,new_c, train['Longitude'], train['Latitude'])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
col_li = ['MedInc', 'Population', 'AveBedrms', 'AveRooms', 'Longitude', 'Latitude']
medinc_ix, pop_ix, brms_ix, rms_ix, log_ix, lat_ix = [train.iloc[:,1:-1].columns.get_loc(c) 
                                                      for c in col_li]

class myFitTrans(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        def cal_dist(a,b,c,x1,y1):
            return abs(a*x1+b*y1+c)/np.sqrt(a**2+b**2)

        medinc_per_pop = X.iloc[:,medinc_ix] / X.iloc[:,pop_ix]
        brms_per_rms = X.iloc[:,brms_ix] / X.iloc[:,rms_ix]
        dist = cal_dist(a,-1,new_c, X.iloc[:,log_ix], X.iloc[:,lat_ix])
        dist = abs()
        return np.c_[X, medinc_per_pop, brms_per_rms, dist]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
pipeline = Pipeline([
        ('adder', myFitTrans(ocean_dist=True)),
        ('std_scaler', StandardScaler())])
poly_pipeline = Pipeline([
        ('adder', myFitTrans(ocean_dist=True)),
        ('poly_features', PolynomialFeatures(degree=2, include_bias=False)),
        ('std_scaler', StandardScaler())])
train_pre = pipeline.fit_transform(train.iloc[:,1:-1])
train_pre_poly = poly_pipeline.fit_transform(train.iloc[:,1:-1])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class myFitTrans(BaseEstimator, TransformerMixin):
    def __init__(self, ocean_dist=True):
        self.ocean_dist = ocean_dist
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X['medinc_per_pop'] = X['MedInc'] / X['Population']
        X['brms_per_rms'] = X['AveBedrms'] / X['AveRooms']
        if self.ocean_dist:    
            a,c = np.polyfit(X['Longitude'], X['Latitude'],1)
            new_c = c-1.8; b=-1
            X['dist'] = abs(a*X['Longitude']+b*X['Latitude']+new_c)/np.sqrt(a**2+b**2)
        return X
    def get_feature_names_out(self, feature_names_in):
        return feature_names_in
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
X_train, y_train = train.drop(columns=['id','MedHouseVal']), train['MedHouseVal']

mypipeline = Pipeline([
        ('adder', myFitTrans(ocean_dist=True)),
        ('std_scaler', StandardScaler()),
        ('RandomForest', RandomForestRegressor())])
mypipeline.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'adder__ocean_dist': [True, False],
              'RandomForest__random_state': [42],
              'RandomForest__max_features': [1,3,5,10,15,20],
              'RandomForest__n_estimators': [5,10,40,65,80,100]}
grid_search = GridSearchCV(mypipeline, param_grid, cv=3, scoring='neg_mean_squared_error', 
                           return_train_score=True, n_jobs=-1, verbose=4, error_score='raise')
grid_search.fit(X_train, y_train)
```
